## silhouette

A) 段定义 + 小工具

```python
import numpy as np

# —— 按你的骨架命名来分段（影响集合用于基于 skin 权重挑选属于该段的顶点。）——
SEG_DEFS = {
    'torso'     : ('RootPart1_M','Chest_M', {'RootPart1_M','Spine1_M','Spine1Part1_M','Chest_M'}),
    'thigh_L'   : ('Hip_L','Knee_L',        {'Hip_L','HipPart1_L','Knee_L'}),
    'calf_L'    : ('Knee_L','Ankle_L',      {'Knee_L','Ankle_L'}),
    'thigh_R'   : ('Hip_R','Knee_R',        {'Hip_R','HipPart1_R','Knee_R'}),
    'calf_R'    : ('Knee_R','Ankle_R',      {'Knee_R','Ankle_R'}),
    'upperarm_L': ('Shoulder_L','Elbow_L',  {'Shoulder_L','ShoulderPart1_L','Elbow_L'}),
    'forearm_L' : ('Elbow_L','Wrist_L',     {'Elbow_L','ElbowPart1_L','Wrist_L'}),
    'upperarm_R': ('Shoulder_R','Elbow_R',  {'Shoulder_R','ShoulderPart1_R','Elbow_R'}),
    'forearm_R' : ('Elbow_R','Wrist_R',     {'Elbow_R','ElbowPart1_R','Wrist_R'}),
}

def _point_segment_dist_and_t(P, A, B):
    """点到线段 AB 的最近距离，同时返回参数 t（投影点在线段上的位置，0 在 A，1 在 B）"""
    v = B - A 
    vv = float(np.dot(v, v)) + 1e-12
    t = float(np.dot(P - A, v) / vv)
    t = max(0.0, min(1.0, t))
    proj = A + t * v
    return float(np.linalg.norm(P - proj)), t
```

B) T-pose：用 per_vertex_weights + V_concat 估计每段世界半径

```python
def precompute_segment_r_world_from_weights(Tpose_globals, V_concat, per_vertex_weights,
    *, quantile=0.80, safety=1.10, weight_thr=0.30, use_cone=True, min_samples=50):
    """
    功能：用 per_vertex_weights + V_concat 在 T-pose 世界坐标 下，为每一段估出“半径”（粗细）。
    V_concat：T-pose 下所有顶点位置（Nx3）。
    per_vertex_weights：每个顶点的 “关节→权重” 字典。
    * 左边的参数可以按位置传参（positional）也可以按名称传参（keyword）
    * 右边的参数只能用关键字传参，不能用位置传参
    quantile（分位数，统计学概念）：用 80% 分位数当半径，比最大值稳健（不受离群点影响）。
    safety：安全系数，估出来的半径再乘一个 >1 的小数，留余量。
    weight_thr：顶点属于某段的“相关权重和”阈值。
    use_cone=True：每段用“锥形半径”（靠近起点半径 r0，靠近终点半径 r1），否则用恒定半径。
    min_samples：如果某段挑出的点太少，退化为“全体顶点”来统计，避免样本过少。
    """
    # 从 T-pose 的关节 4x4 变换里取平移列（世界坐标），当作“线段”两端。
    seg_lines = {}
    for seg, (j0, j1, _) in SEG_DEFS.items():
        if j0 in Tpose_globals and j1 in Tpose_globals:
            A = Tpose_globals[j0][:3, 3]
            B = Tpose_globals[j1][:3, 3]
            seg_lines[seg] = (A, B)
    # 一一对应：第 i 个顶点的位置和权重要对齐。
    N = len(V_concat)
    assert N == len(per_vertex_weights), "V_concat 与 per_vertex_weights 长度不一致！"

    radii = {} # 最终返回：每段 → 半径参数
    all_idx = np.arange(N) # 备选：挑不到足够多的点时兜底。

    for seg, (A, B) in seg_lines.items():
        jset = SEG_DEFS[seg][2] #当前段的“相关关节集合”（比如大腿：Hip/Thigh/Knee）。

        # —— 选出该段的候选顶点（基于权重）——
        idx = []
        for i, wdict in enumerate(per_vertex_weights):
            if not wdict: 
                continue
            sw = sum(float(wdict.get(j, 0.0)) for j in jset)  # 相关关节权重和
            if sw >= weight_thr:
                idx.append(i); continue
            # 最大权重归属作为兜底
            jmax = max(wdict.items(), key=lambda kv: kv[1])[0]
            if jmax in jset:
                idx.append(i)

        cand = np.array(idx, dtype=int) if idx else all_idx

        # 计算这些点到线段的距离分布，对候选顶点，计算它们到段的最近距离 d，和落点位置 t（[0,1]）。
        P = V_concat[cand]
        dists, ts = [], []
        for p in P:
            d, t = _point_segment_dist_and_t(p, A, B)
            dists.append(d); ts.append(t)
        dists = np.asarray(dists); ts = np.asarray(ts)

        if len(dists) < min_samples:
            # 再兜底：全体
            P = V_concat
            dists, ts = [], []
            for p in P:
                d, t = _point_segment_dist_and_t(p, A, B)
                dists.append(d); ts.append(t)
            dists = np.asarray(dists); ts = np.asarray(ts)

        eps = 1e-4
        if use_cone:
            left  = dists[ts < 0.5] if (ts < 0.5).any() else dists
            right = dists[ts >= 0.5] if (ts >= 0.5).any() else dists
            r0 = float(np.quantile(left,  quantile) * safety)
            r1 = float(np.quantile(right, quantile) * safety)
            radii[seg] = {'type':'cone','r0':max(r0,eps),'r1':max(r1,eps)}
        else:
            r = float(np.quantile(dists, quantile) * safety)
            radii[seg] = {'type':'capsule','r':max(r,eps)}
    return radii
```

 运行时：世界半径 → 像素半径；构建 2D 胶囊

```python
def _world_to_cam(xw, M_cam_to_world):
    Rcw = M_cam_to_world[:3, :3]; tcw = M_cam_to_world[:3, 3]
    Xc = Rcw.T @ (xw - tcw)
    return np.array([Xc[0], Xc[1], -Xc[2]])  # z>0 表示在相机前方（配合你项目的 -Z 摄像机）

def _world_radius_to_px(r_world, z_cam, K):
    fx = float(K[0,0])
    return fx * float(r_world) / max(z_cam, 1e-6)

def capsule_r_px_runtime(pose_globals_now, K, M_cam_to_world, radii_world):
    "运行时从当前姿势的全局矩阵里取线段端点 A/B；算它们到相机的深度 zA,zB（z>0 才在前方）。"
    r_px = {}
    for seg, (j0,j1,_) in SEG_DEFS.items():
        if seg not in radii_world: 
            continue
        if j0 not in pose_globals_now or j1 not in pose_globals_now:
            continue
        A = pose_globals_now[j0][:3,3]
        B = pose_globals_now[j1][:3,3]
        zA = _world_to_cam(A, M_cam_to_world)[2]
        zB = _world_to_cam(B, M_cam_to_world)[2]
        if zA <= 0 or zB <= 0:
            continue
        #把世界半径换算成像素半径。锥形就两头分别换算，取大值“保险”；常半径用中间深度近似。
        rw = radii_world[seg]
        if rw['type'] == 'cone':
            rA = _world_radius_to_px(rw['r0'], zA, K)
            rB = _world_radius_to_px(rw['r1'], zB, K)
            r_px[seg] = float(max(rA, rB))   # 安全取大
        else:
            r_mid = _world_radius_to_px(rw['r'], 0.5*(zA+zB), K)
            r_px[seg] = float(r_mid)
    return r_px

def build_capsules_px(points_2d, r_px_by_seg):
    "把 2D 关节点连成“线段”，配上像素半径，得到 2D 胶囊（线段+半径）。"
    "肘部也给一个小半径的“点胶囊”（简化肘尖遮挡判断）。"
    C = {}
    def _uv(j): 
        return np.array(points_2d[j], dtype=float) if j in points_2d else None
    for seg, (j0,j1,_) in SEG_DEFS.items():
        if seg not in r_px_by_seg: 
            continue
        p0 = _uv(j0); p1 = _uv(j1)
        if p0 is None or p1 is None:
            continue
        C[seg] = dict(p0=p0, p1=p1, r=float(r_px_by_seg[seg]))
    # 可选：额外加入肘部点（用前臂胶囊也行）
    if 'Elbow_L' in points_2d:
        C['elbow_L_pt'] = dict(pt=np.array(points_2d['Elbow_L'], float), r=r_px_by_seg.get('forearm_L', 6.0))
    if 'Elbow_R' in points_2d:
        C['elbow_R_pt'] = dict(pt=np.array(points_2d['Elbow_R'], float), r=r_px_by_seg.get('forearm_R', 6.0))
    return C
```

2D 胶囊几何 & silhouette 评分

```python
# —— 2D 几何 —— 
def _clamp01(x): return max(0.0, min(1.0, x))

def _seg_seg_min_dist(p, q, r, s):
    "功能：两条 2D 线段的最小距离。"
    u = q - p; v = s - r; w0 = p - r
    a = np.dot(u,u) + 1e-12
    b = np.dot(u,v)
    c = np.dot(v,v) + 1e-12
    d = np.dot(u,w0)
    e = np.dot(v,w0)
    D = a*c - b*b
    sc = _clamp01((b*e - c*d)/D) if D>1e-12 else 0.0
    tc = _clamp01((a*e - b*d)/D) if D>1e-12 else 0.0
    cp = p + sc*u; cq = r + tc*v
    return np.linalg.norm(cp - cq)

def capsule_minsep(capA, capB):
    "点到胶囊的“边界最小间距”。同理：距离 − 半径。"
    d = _seg_seg_min_dist(capA['p0'], capA['p1'], capB['p0'], capB['p1'])
    return d - (capA['r'] + capB['r'])

def point_capsule_minsep(pt, cap):
    p, q = cap['p0'], cap['p1']
    v = q - p; t = np.dot(pt - p, v) / (np.dot(v,v) + 1e-12)
    t = _clamp01(t); closest = p + t*v
    d = np.linalg.norm(pt - closest)
    return d - cap['r']

# —— 新增：统一比较任意两种（线 vs 线、点 vs 线、点 vs 点）
def _minsep_any(itemA, itemB):
    # item 可能是 {'p0','p1','r'} 或 {'pt','r'}
    is_ptA = ('pt' in itemA); is_ptB = ('pt' in itemB)
    if not is_ptA and not is_ptB:
        return capsule_minsep(itemA, itemB)
    if is_ptA and not is_ptB:
        return point_capsule_minsep(itemA['pt'], itemB)
    if not is_ptA and is_ptB:
        return point_capsule_minsep(itemB['pt'], itemA)
    # pt vs pt
    d = np.linalg.norm(itemA['pt'] - itemB['pt'])
    return d - (itemA['r'] + itemB['r'])

# —— 新增：两组名字之间取最小间距 + 记录是谁最小
def _minsep_groups(C, namesA, namesB):
    best = None; arg = None
    for a in namesA:
        if a not in C: 
            continue
        for b in namesB:
            if b not in C: 
                continue
            s = _minsep_any(C[a], C[b])
            if (best is None) or (s < best):
                best, arg = s, (a, b)
    return best, arg  # best 可能是 None（两组都缺）

# 把“要比较的名字组”组织起来 我们把每条腿/手臂涉及的胶囊名列出来（上臂、前臂、肘点等都纳入），这样写扩展性强：
def _names_leg(side):  # 'L' or 'R'
    return [f'thigh_{side}', f'calf_{side}']

def _names_arm(side):  # 'L' or 'R'
    # 既考虑线胶囊，也考虑肘点（此前 build_capsules_px 已经加了 elbow_*_pt）
    return [f'upperarm_{side}', f'forearm_{side}', f'elbow_{side}_pt']


# —— 选 LoA 腿侧（你的 LoA 输出即可）——
def loa_leg_side(selected_loa):
    "看 LoA 是否包含左/右腿关节；如果只包含某一侧，说明那一侧是“主腿”，另一侧优先拿去分离（减少破坏 LoA）。"
    names = [j for j,_ in selected_loa] if selected_loa else []
    if any(n in names for n in ('Knee_L','Ankle_L')) and not any(n in names for n in ('Knee_R','Ankle_R')):
        return 'L'
    if any(n in names for n in ('Knee_R','Ankle_R')) and not any(n in names for n in ('Knee_L','Ankle_L')):
        return 'R'
    return None  # trunk-only 或两腿都有

# —— 评分 —— 
def gap_reward(minsep, target_px=12.0, overlap_penalty=2.0):
    """目标是希望“最小间距 ≥ target_px”。
    如果间距是负（重叠），给负分，最多扣到 -overlap_penalty。
    间距正，按比例涨分，上限 1。（把几何量转成 0~1 的奖励）。"""
    if minsep < 0:
        return max(-overlap_penalty, minsep/(-target_px))
    return min(1.0, minsep/target_px)

def silhouette_terms_from_capsules(C):
    torso = C.get('torso', None)
    if torso is None:
        return dict(legL=None, legR=None, elbowL=None, elbowR=None)

    def leg_side(side):
        thigh = C.get(f'thigh_{side}'); calf = C.get(f'calf_{side}')
        vals = []
        if thigh is not None: vals.append(capsule_minsep(thigh, torso))
        if calf  is not None: vals.append(capsule_minsep(calf,  torso))
        return min(vals) if vals else None

    legL = leg_side('L'); legR = leg_side('R')

    def elbow_side(side):
        fore = C.get(f'forearm_{side}')
        if fore is not None:
            return capsule_minsep(fore, torso)
        pt = C.get(f'elbow_{side}_pt', None)
        if pt is None: return None
        return point_capsule_minsep(pt['pt'], torso)

    elbowL = elbow_side('L'); elbowR = elbow_side('R')
    return dict(legL=legL, legR=legR, elbowL=elbowL, elbowR=elbowR)

def silhouette_terms_from_capsules_v2(C):
    terms  = {}
    argmin = {}

    # 躯干在不在
    torso_names = ['torso'] if ('torso' in C) else []

    # —— 躯干 vs 腿（左右）——
    for s in ('L', 'R'):
        names_leg = _names_leg(s)
        val, arg = _minsep_groups(C, torso_names, names_leg)
        terms[f'torso_leg_{s}'] = val
        argmin[f'torso_leg_{s}'] = arg

    # —— 躯干 vs 手臂（左右）——
    for s in ('L', 'R'):
        names_arm = _names_arm(s)
        val, arg = _minsep_groups(C, torso_names, names_arm)
        terms[f'torso_arm_{s}'] = val
        argmin[f'torso_arm_{s}'] = arg

    # —— 腿 vs 腿（L–R，取跨侧的最小）——
    val, arg = _minsep_groups(C, _names_leg('L'), _names_leg('R'))
    terms['leg_leg'] = val
    argmin['leg_leg'] = arg

    # —— 手臂 vs 手臂（L–R）——
    val, arg = _minsep_groups(C, _names_arm('L'), _names_arm('R'))
    terms['arm_arm'] = val
    argmin['arm_arm'] = arg

    # —— 手臂 vs 腿（跨侧更常见：左臂 vs 右腿、右臂 vs 左腿）——
    v1, a1 = _minsep_groups(C, _names_arm('L'), _names_leg('R'))
    v2, a2 = _minsep_groups(C, _names_arm('R'), _names_leg('L'))
    # 取两种跨侧匹配中的更小者
    if (v1 is None) and (v2 is None):
        terms['arm_leg'] = None; argmin['arm_leg'] = None
    elif (v2 is None) or (v1 is not None and v1 <= v2):
        terms['arm_leg'] = v1;   argmin['arm_leg'] = a1
    else:
        terms['arm_leg'] = v2;   argmin['arm_leg'] = a2

    return terms, argmin
    
def silhouette_score(points_2d, r_px_by_seg, keep_ctrl=None, selected_loa=None, 
                     target_gap_px=10, w_leg_torso=0.7, w_elbow_torso=0.0, w_keep=0.3,
                     w_arm_torso=0.15, w_leg_leg=0.15, w_arm_arm=0.15, w_arm_leg=0.15,
                     tgt_leg_torso=None, tgt_arm_torso=None, tgt_leg_leg=None, tgt_arm_arm=None, 
                     tgt_arm_leg=None):
    """
    target_gap_px: 全局默认“希望的最小缝宽”（像素）。镜头近可大些（12–18），远景小些（6–10）。
    w_leg_torso: “腿 vs 躯干”这项的权重（主力项，默认 0.7）。只对被选中可动的那条腿计分（优先动非 LoA 那条腿）。
    w_elbow_torso: “肘 vs 躯干”的权重，保持 0 就行。
    w_keep: “保持 LoA 走向”的权重（用你的 keep_ctrl 做参照）。越大越不容易把黄线扯坏。
    w_arm_torso: “手臂（上臂/前臂/肘点）vs 躯干”的权重。手臂老贴身就加一点（默认 0.15，建议 0.1–0.3）。
    w_leg_leg: “腿 vs 腿（左右腿互相）”的权重。两腿重叠多时加（默认 0.15）。
    w_arm_arm: “手臂 vs 手臂（左右臂互相）”的权重。抱臂以外不想贴就加（默认 0.15）。
    w_arm_leg: “手臂 vs 腿（跨侧优先，如左臂-右腿）”的权重。常见交叉遮挡时加（默认 0.15）。
    tgt_leg_torso / tgt_arm_torso / tgt_leg_leg / tgt_arm_arm / tgt_arm_leg: 各类别的目标缝宽；不填时按默认比例：
    leg_torso：= target_gap_px
    arm_torso：≈ 0.7 * target_gap_px
    leg_leg：≈ 0.8 * target_gap_px
    arm_arm：≈ 0.7 * target_gap_px
    arm_leg：≈ 0.8 * target_gap_px
    """
    # 目标间距：不传就用主 target
    tgt_leg_torso = target_gap_px if tgt_leg_torso is None else tgt_leg_torso
    tgt_arm_torso = 0.7*target_gap_px if tgt_arm_torso is None else tgt_arm_torso
    tgt_leg_leg   = 0.8*target_gap_px if tgt_leg_leg   is None else tgt_leg_leg
    tgt_arm_arm   = 0.7*target_gap_px if tgt_arm_arm   is None else tgt_arm_arm
    tgt_arm_leg   = 0.8*target_gap_px if tgt_arm_leg   is None else tgt_arm_leg

    C = build_capsules_px(points_2d, r_px_by_seg)
    terms, argmins = silhouette_terms_from_capsules_v2(C)

    # —— 仍按“动非 LoA 腿”的策略，选择 leg_torso 的那一侧计分 —— 
    move = loa_leg_side(selected_loa)
    if move is None:
        cand = { 'L': terms['torso_leg_L'], 'R': terms['torso_leg_R'] }
        exist = [(k,v) for k,v in cand.items() if v is not None]
        move = (min(exist, key=lambda kv: kv[1])[0]) if exist else None
    else:
        move = ('R' if move=='L' else 'L')

    sc_leg_torso = 0.0
    if move == 'L' and terms['torso_leg_L'] is not None:
        sc_leg_torso = gap_reward(terms['torso_leg_L'], tgt_leg_torso)
    elif move == 'R' and terms['torso_leg_R'] is not None:
        sc_leg_torso = gap_reward(terms['torso_leg_R'], tgt_leg_torso)

    # —— 其他轻权重项：只要存在就加一点奖励/惩罚 —— 
    sc_arm_torso = 0.0
    for s in ('L','R'):
        v = terms[f'torso_arm_{s}']
        if v is not None:
            sc_arm_torso += 0.5 * gap_reward(v, tgt_arm_torso)  # 左右各占 0.5

    sc_leg_leg = 0.0 if terms['leg_leg'] is None else gap_reward(terms['leg_leg'], tgt_leg_leg)
    sc_arm_arm = 0.0 if terms['arm_arm'] is None else gap_reward(terms['arm_arm'], tgt_arm_arm)
    sc_arm_leg = 0.0 if terms['arm_leg'] is None else gap_reward(terms['arm_leg'], tgt_arm_leg)

    sc_keep = 0.0
    if keep_ctrl is not None and selected_loa:
        import score
        sc_keep, _, _ = score.line_of_action_score(selected_loa, keep_ctrl)

    score_total = (w_leg_torso  * sc_leg_torso +
                   w_arm_torso  * sc_arm_torso +
                   w_leg_leg    * sc_leg_leg   +
                   w_arm_arm    * sc_arm_arm   +
                   w_arm_leg    * sc_arm_leg   +
                   w_keep       * sc_keep)

    return float(score_total), terms, argmins, move
```

```python
import copy

# ---------- gene 顺序 ----------
OPT_JOINTS =['Hip_L','Hip_R','RootPart1_M','Spine1_M','Knee_L','Knee_R','Spine1Part1_M',
            'Ankle_L','Ankle_R','Chest_M','Neck_M','Shoulder_L','Shoulder_R','NeckPart1_M',
             'Head_M','Elbow_L','Elbow_R','Wrist_L','Wrist_R'] # 列表

FROZEN_ARMS = {'Shoulder_L','Shoulder_R','Elbow_L','Elbow_R','Wrist_L','Wrist_R'} # 集合
arms_only = {'Shoulder_L','Shoulder_R'}
legs_only = {'Hip_L','Hip_R'}

# ---------- 只有 Z 轴可旋转的关节 ----------
Z_ONLY_leg = {'Knee_L','Knee_R'} # 集合
Z_ONLY_arm = {'Elbow_L','Elbow_R'}

def stageA_param_ranges():
    pr = []
    for jn in OPT_JOINTS:
        if jn in FROZEN_ARMS:
            pr += [(0, 0), (0, 0), (0, 0)]
        elif jn in Z_ONLY_leg:
            pr += [(0, 0), (0, 0), (-10, 10)]
        else:
            pr += [(-30, 30), (-30, 30), (-30, 30)]
    return pr

def stageB_param_ranges():
    pr = []
    for jn in OPT_JOINTS:
        if jn == 'Head_M':
            pr += [(-45, 45), (-45, 45), (-45, 45)]
        else:
            pr += [(0, 0), (0, 0), (0, 0)]
    return pr

def stageC_param_ranges():
    pr = []
    for jn in OPT_JOINTS:
        if jn in Z_ONLY_leg or Z_ONLY_arm:
            pr += [(0, 0), (0, 0), (-10, 10)]
        elif jn in arms_only or legs_only:
            pr += [(-10, 10), (-10, 10), (-10, 10)]
        else:
            pr += [(0, 0), (0, 0), (0, 0)]
    return pr
    
joint_idx_map = {jn: i*3 for i, jn in enumerate(OPT_JOINTS)}
```

```python
import score
import numpy as np

def ranges_to_mask(param_ranges):
    # 0/1 掩码：范围为 (0,0) 的位置置 0
    m = []
    for lo, hi in param_ranges:
        m.append(0.0 if (lo == 0 and hi == 0) else 1.0)
    return np.array(m, dtype=float)

def evaluate_pose(vector, stage='A', joint_order=OPT_JOINTS, base_vec=None, base_mask=None, param_ranges=None):  

    # —— 组合向量：base_vec + mask * vector ——
    vec = np.array(vector, dtype=float)
    if base_vec is None:
        composed = vec
    else:
        if base_mask is None:
            base_mask = np.ones_like(vec)
        composed = np.array(base_vec, float) + np.array(base_mask, float) * vec
        
    # 1) 深拷贝姿势，并可选Root平移
    pose_JR = copy.deepcopy(LoAfix_poseJR)

    # 2) 叠加各关节旋转增量
    for jn in joint_order:
        idx = joint_idx_map[jn]
        dx, dy, dz = vector[idx:idx+3]
        ox, oy, oz = pose_JR["joint_rot"][jn]
        pose_JR["joint_rot"][jn] = [ox+dx, oy+dy, oz+dz]

    # 3) 构建全局 & loa准备
    pose_globals = PP.build_pose_globals(sorted_entries, local_bind, parent_map, pose_JR, 'Root_M')
    points_2d    = LOAO.project_pose_to_2d(pose_globals, K, M_cam_to_world) #计算loa评分
    _, selected_loa = LOAO.extract_loa_candidates(points_2d, align_max_deg=35, knee_straight_min_deg=135)

    # 4) Stage 专属评分
    if stage == 'A':
        loa_sc, _, _ = score.line_of_action_score(selected_loa, best_ctrl)  # LoA
        fitness = (1.0 * loa_sc)
        return (float(fitness),)

    elif stage == 'B':
        cam_head = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")
        const_head = score.rigid_penalty_global_v2(pose_globals, pose_globals_orig,
                                        constraints={
                                            'HeadEnd_M':  dict(w_t=0.005, w_r=0),
                                            'Wrist_L': dict(w_t=0, w_r=0),
                                            'Wrist_R': dict(w_t=0, w_r=0),
                                            'Ankle_L': dict(w_t=0, w_r=0),
                                            'Ankle_R': dict(w_t=0, w_r=0),},
                                        use_toe_anchor=True)   # 位置旋转约束
        fitness = (0.8 * cam_head + 0.2 * const_head)
        return (float(fitness),)

    elif stage == 'C':
        r_px_by_seg    = capsule_r_px_runtime(pose_globals, K, M_cam_to_world, radii_world) # 运行时半径
        sc_sil, sil_terms, sil_argmins, move_leg = silhouette_score(points_2d, r_px_by_seg,
                                                    keep_ctrl=best_ctrl, selected_loa=selected_loa,
                                                    target_gap_px=10.0,
                                                    w_leg_torso=0.7, w_arm_torso=0.15, w_leg_leg=0.15, 
                                                    w_arm_arm=0.15, w_arm_leg=0.15, w_keep=0.3)
        fitness = (1.0 * sc_sil)
        return (float(fitness),)
```

```python
from functools import partial
from deap import creator, tools, algorithms
import deap.base as deap_base
import random, copy
import numpy as np

def make_mutGaussian_clamp(param_ranges):
    def _mut(individual, mu, sigma, indpb):
        # 高斯变异
        tools.mutGaussian(individual, mu, sigma, indpb)
        # clamp
        for i, (low, high) in enumerate(param_ranges):
            if individual[i] < low:   individual[i] = low
            elif individual[i] > high: individual[i] = high
        return (individual,)
    return _mut
```

```python
def optimize_pose_GA(eval_fn, param_ranges,
                     ngen=40, mu=60, lambda_=120, cxpb=0.5, mutpb=0.2,
                     sigma=5.0, seed_vec=None, seed_spread=2.0):
    """
    ngen: 进化的代数（generations）。运行 N 代，每代都会进行选择、交叉、变异并生成新一批个体。
        越大 → 计算时间长，但可能找到更优解。
    mu: 种群规模（population size），也就是每一代保留多少个个体。这里用的是 eaMuPlusLambda 算法，
        所以 mu 是精英个体的数量。越大 → 覆盖搜索空间更广，但计算更慢。
    lambda_: 每代新产生的子代数量。在 eaMuPlusLambda 里：从 mu 个体中选父代,交叉 / 变异生成 lambda_ 个新个体,
        再和原有的 mu 个体一起挑选下一代。
    cxpb: 交叉概率（crossover probability）。决定了两两个个体交叉生成新个体的概率。0.5 表示 50% 的概率交叉，
        50% 的概率直接复制父代。
    mutpb: 变异概率（mutation probability）。决定了每个个体在交叉后会被变异的概率。0.2 = 20% 概率变异。
    sigma: 变异的标准差（标准差越大，变异幅度越大）。这里用的 mutGaussian 高斯变异，所以 sigma 控制随机扰动的强度。
        例如参数值是 30，sigma=5.0 → 变异后可能是 30 ± 一些随机数。
    seed_vec: 种群初始化时的引导向量（上一阶段的最佳解）。如果有值，就会在它附近随机生成个体（而不是完全随机）。
        好处是热启动，加快收敛。
    seed_spread: 初始化时在 seed 附近加的高斯扰动强度。越大 → 初始种群更分散，探索范围大；越小 → 初始种群更集中，收敛快。
    """
    # 定义适应度和个体
    if 'FitnessMax' in creator.__dict__:
        try:
            del creator.FitnessMax; del creator.Individual
        except Exception:
            pass
    creator.create('FitnessMax', deap_base.Fitness, weights=(1.0,))
    creator.create('Individual', list, fitness=creator.FitnessMax)

    toolbox = deap_base.Toolbox()

    # —— 初始化个体：支持基于 seed 的偏置初始化 ——
    dim = len(param_ranges)
    lows  = np.array([a for a,_ in param_ranges], dtype=float)
    highs = np.array([b for _,b in param_ranges], dtype=float)

    def _rand_ind():
        return [random.uniform(lo, hi) for lo,hi in param_ranges]#我靠 误我好久阿我靠

    def _zero_ind():
        return [0.0 for _ in param_ranges]

    if seed_vec is None:
        init_fn = _zero_ind
    else:
        seed = np.array(seed_vec, dtype=float)
        def _biased():
            cand = seed + np.random.normal(0.0, seed_spread, size=dim)
            cand = np.clip(cand, lows, highs)
            return cand.tolist()
        init_fn = _biased

    toolbox.register('individual', tools.initIterate, creator.Individual, init_fn)
    toolbox.register('population', tools.initRepeat, list, toolbox.individual)

    # —— 注册算子（注意把 param_ranges 注入 mutate） ——
    toolbox.register('evaluate', eval_fn)
    toolbox.register('mate',     tools.cxBlend, alpha=0.4)
    toolbox.register('mutate',   make_mutGaussian_clamp(param_ranges), mu=0, sigma=sigma, indpb=0.2)
    toolbox.register('select',   tools.selTournament, tournsize=3)

    pop = toolbox.population(mu)
    algorithms.eaMuPlusLambda(pop, toolbox,
                              mu       = mu,
                              lambda_  = lambda_,
                              cxpb     = cxpb,
                              mutpb    = mutpb,
                              ngen     = ngen,
                              verbose  = False)
    best = tools.selBest(pop, k=1)[0]
    return best
```

```python
# —— Stage A ——
prA = stageA_param_ranges()
fnA = lambda v: evaluate_pose(v, stage='A', param_ranges=prA)
bestA = optimize_pose_GA(fnA, prA, ngen=20, mu=30, lambda_=60, sigma=5.0, seed_vec=None)

# —— Stage B ——
prB = stageB_param_ranges()
maskB = ranges_to_mask(prB)
fnB = lambda v: evaluate_pose(v, stage='B', base_vec=bestA, base_mask=maskB, param_ranges=prB)
bestB = optimize_pose_GA(fnB, prB, ngen=20, mu=30, lambda_=60, mutpb=0.4, sigma=5.0, seed_vec=np.zeros(len(prB)), seed_spread=1.0)

# —— Stage C ——
prC = stageC_param_ranges()
maskC = ranges_to_mask(prC)
fnC = lambda v: evaluate_pose(v, stage='C', base_vec=bestB, base_mask=maskC, param_ranges=prC)
bestC = optimize_pose_GA(fnC, prC, ngen=20, mu=30, lambda_=60, sigma=1.5, seed_vec=np.zeros(len(prC)), seed_spread=0.7)
```
