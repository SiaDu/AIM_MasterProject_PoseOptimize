```python
import score
import numpy as np

def ranges_to_mask(param_ranges):
    # 0/1 掩码：范围为 (0,0) 的位置置 0
    m = []
    for lo, hi in param_ranges:
        m.append(0.0 if (lo == 0 and hi == 0) else 1.0)
    return np.array(m, dtype=float)

def evaluate_pose(vector, stage='A', joint_order=OPT_JOINTS):
    """
    stage: 'A' | 'B' | 'C'
      A：LoA + const（可用RootYaw等大自由度）
      B：身体朝向 + 平滑（锁Root，只微调躯干）
      C：仅Head_M对相机；腿部小幅分离；保持LoA走向
    """

    # 1) 深拷贝姿势，并可选Root平移
    pose_JR = copy.deepcopy(LoAfix_poseJR)

    # 2) 叠加各关节旋转增量
    for jn in joint_order:
        idx = joint_idx_map[jn]
        dx, dy, dz = vector[idx:idx+3]
        ox, oy, oz = pose_JR["joint_rot"][jn]
        pose_JR["joint_rot"][jn] = [ox+dx, oy+dy, oz+dz]

    # 3) 构建全局 & loa准备
    pose_globals = PP.build_pose_globals(sorted_entries, local_bind, parent_map, pose_JR, 'Root_M')
    points_2d    = LOAO.project_pose_to_2d(pose_globals, K, M_cam_to_world) #计算loa评分
    _, selected_loa = LOAO.extract_loa_candidates(points_2d, align_max_deg=35, knee_straight_min_deg=135)

    # 4) Stage 专属评分
    if stage == 'A':
        loa_sc, _, _ = score.line_of_action_score(selected_loa, best_ctrl)  # LoA
        # const = score.rigid_penalty_global_v2(pose_globals, pose_globals_orig,
        #                                 constraints={
        #                                     'HeadEnd_M':  dict(w_t=0.002, w_r=0),
        #                                     'Wrist_L': dict(w_t=0, w_r=0),
        #                                     'Wrist_R': dict(w_t=0, w_r=0),
        #                                     'Ankle_L': dict(w_t=0.005, w_r=0.02),
        #                                     'Ankle_R': dict(w_t=0.005, w_r=0.02),},
        #                                 use_toe_anchor=True)   # 位置旋转约束
        # smth = score.smooth_penalty(pose_JR['joint_rot'], LoAfix_poseJR['joint_rot'], lam=1e-3)#相对原姿最小变化
        # cam_body = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Chest_M")
        # cam_head = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")#面朝相机
        # sil_leg  = silhouette_leg_gap(points_2d, cap=15.0, w=0.3)
        # 组合权重：LoA 主导，约束为辅
        fitness = (1.0 * loa_sc)
        return (float(fitness),)

    elif stage == 'B':
        cam_head = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")
        const_head = score.rigid_penalty_global_v2(pose_globals, pose_globals_orig,
                                        constraints={
                                            'HeadEnd_M':  dict(w_t=0.005, w_r=0),
                                            'Wrist_L': dict(w_t=0, w_r=0),
                                            'Wrist_R': dict(w_t=0, w_r=0),
                                            'Ankle_L': dict(w_t=0, w_r=0),
                                            'Ankle_R': dict(w_t=0, w_r=0),},
                                        use_toe_anchor=True)   # 位置旋转约束
        fitness = (1.0 * cam_head +
                   0 * const_head)
        return (float(fitness),)

    # elif stage == 'C':
    #     # --- 只改 Head_M；腿部小幅分离；保持 LoA ---
    #     cam_head = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")
    #     #sil_leg  = silhouette_leg_gap(points_2d, cap=15.0, w=0.3)
    #     keeploa_sc, _, _ = score.line_of_action_score(selected_loa, best_ctrl)
    #     fitness = (1.0 * cam_head +
    #                0.5 * sil_leg +
    #                1.0 * keeploa_sc)
    #     return (float(fitness),)
```

```python
from functools import partial
from deap import creator, tools, algorithms
import deap.base as deap_base
import random, copy
import numpy as np

def make_mutGaussian_clamp(param_ranges):
    def _mut(individual, mu, sigma, indpb):
        # 高斯变异
        tools.mutGaussian(individual, mu, sigma, indpb)
        # clamp
        for i, (low, high) in enumerate(param_ranges):
            if individual[i] < low:   individual[i] = low
            elif individual[i] > high: individual[i] = high
        return (individual,)
    return _mut
```

```python
def optimize_pose_GA(eval_fn, param_ranges,
                     ngen=40, mu=60, lambda_=120, cxpb=0.5, mutpb=0.2,
                     sigma=5.0, seed_vec=None, seed_spread=2.0):
    """
    ngen: 进化的代数（generations）。运行 N 代，每代都会进行选择、交叉、变异并生成新一批个体。
        越大 → 计算时间长，但可能找到更优解。
    mu: 种群规模（population size），也就是每一代保留多少个个体。这里用的是 eaMuPlusLambda 算法，
        所以 mu 是精英个体的数量。越大 → 覆盖搜索空间更广，但计算更慢。
    lambda_: 每代新产生的子代数量。在 eaMuPlusLambda 里：从 mu 个体中选父代,交叉 / 变异生成 lambda_ 个新个体,
        再和原有的 mu 个体一起挑选下一代。
    cxpb: 交叉概率（crossover probability）。决定了两两个个体交叉生成新个体的概率。0.5 表示 50% 的概率交叉，
        50% 的概率直接复制父代。
    mutpb: 变异概率（mutation probability）。决定了每个个体在交叉后会被变异的概率。0.2 = 20% 概率变异。
    sigma: 变异的标准差（标准差越大，变异幅度越大）。这里用的 mutGaussian 高斯变异，所以 sigma 控制随机扰动的强度。
        例如参数值是 30，sigma=5.0 → 变异后可能是 30 ± 一些随机数。
    seed_vec: 种群初始化时的引导向量（上一阶段的最佳解）。如果有值，就会在它附近随机生成个体（而不是完全随机）。
        好处是热启动，加快收敛。
    seed_spread: 初始化时在 seed 附近加的高斯扰动强度。越大 → 初始种群更分散，探索范围大；越小 → 初始种群更集中，收敛快。
    """
    # 定义适应度和个体
    if 'FitnessMax' in creator.__dict__:
        try:
            del creator.FitnessMax; del creator.Individual
        except Exception:
            pass
    creator.create('FitnessMax', deap_base.Fitness, weights=(1.0,))
    creator.create('Individual', list, fitness=creator.FitnessMax)

    toolbox = deap_base.Toolbox()

    # —— 初始化个体：支持基于 seed 的偏置初始化 ——
    dim = len(param_ranges)
    lows  = np.array([a for a,_ in param_ranges], dtype=float)
    highs = np.array([b for _,b in param_ranges], dtype=float)

    def _rand_ind():
        return [random.uniform(lo, hi) for lo,hi in param_ranges]#我靠 误我好久阿我靠

    def _zero_ind():
        return [0.0 for _ in param_ranges]

    if seed_vec is None:
        init_fn = _zero_ind
    else:
        seed = np.array(seed_vec, dtype=float)
        def _biased():
            cand = seed + np.random.normal(0.0, seed_spread, size=dim)
            cand = np.clip(cand, lows, highs)
            return cand.tolist()
        init_fn = _biased

    toolbox.register('individual', tools.initIterate, creator.Individual, init_fn)
    toolbox.register('population', tools.initRepeat, list, toolbox.individual)

    # —— 注册算子（注意把 param_ranges 注入 mutate） ——
    toolbox.register('evaluate', eval_fn)
    toolbox.register('mate',     tools.cxBlend, alpha=0.4)
    toolbox.register('mutate',   make_mutGaussian_clamp(param_ranges), mu=0, sigma=sigma, indpb=0.2)
    toolbox.register('select',   tools.selTournament, tournsize=3)

    pop = toolbox.population(mu)
    algorithms.eaMuPlusLambda(pop, toolbox,
                              mu       = mu,
                              lambda_  = lambda_,
                              cxpb     = cxpb,
                              mutpb    = mutpb,
                              ngen     = ngen,
                              verbose  = False)
    best = tools.selBest(pop, k=1)[0]
    return best
```

```python
# —— Stage A ——
prA = stageA_param_ranges()
fnA = lambda v: evaluate_pose(v, stage='A')
bestA = optimize_pose_GA(fnA, prA, ngen=20, mu=30, lambda_=60, sigma=5.0, seed_vec=None)
pose_after_A = apply_vec_to_pose(LoAfix_poseJR, bestA)
# —— Stage B ——
prB = stageB_param_ranges()
fnB = lambda v: evaluate_pose(v, stage='B')
bestB = optimize_pose_GA(fnB, prB, ngen=20, mu=30, lambda_=60, sigma=5.0, seed_vec=bestA, seed_spread=1.0)

# # —— Stage C ——
# prC = stageC_param_ranges()
# fnC = lambda v: evaluate_pose(v, stage='C')
# bestC = optimize_pose_GA(fnC, prC, ngen=20, mu=30, lambda_=60, sigma=1.5, seed_vec=bestB, seed_spread=0.7)
```
