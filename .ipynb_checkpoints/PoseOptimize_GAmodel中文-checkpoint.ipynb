## 2. 载入 T‑pose 绑定数据 & 线性蒙皮（LBS）准备

这里调用两个核心工具模块：

(A) `T_poseFromMayaProcess.load_local_bind_and_orient`, 读取 `joints_hierarchy_local.json`，解析出：
- `parent_map`: 每个关节的父子关系（短名）。
- `sorted_entries`: 按层级深度排序的关节列表（便于自顶向下递归）。
- `local_bind`: 各关节**本地绑定矩阵** $L_j \in \mathbb{R}^{4\times 4}$，列主序读取。

  对于任一关节的**全局绑定矩阵** $G^{\mathrm{bind}}_j$，可通过递归计算：
    
    $$
    G^{\mathrm{bind}}_j =
    \begin{cases}
    L_j, & \text{若 } \mathrm{parent}(j) = \varnothing \\
    G^{\mathrm{bind}}_{\mathrm{parent}(j)} \cdot L_j, & \text{否则}
    \end{cases}
    $$
    
    代码中对应 `compute_bind_global(j)`。

(B) `process_pose.skin_character`, 取 `.../weight/` 目录下导出的：
- `verts_rest_*.json`：各 mesh 的 **rest pose 顶点** $V_{\mathrm{rest}}$。
- `influences_*.json`：各 mesh 的 **关节列表与顺序**（与权重列对齐）。
- `weights_*.json`：**每顶点权重** $W \in \mathbb{R}^{V \times J}$。
- `bind_matrices.json`：**全局绑定矩阵**，并在加载时取逆得到 $B_j^{-1}$。

    给定任一姿态下的**关节世界矩阵** $T_j$（这里使用绑定姿态 $T_j = G^{\mathrm{bind}}_j$ 做校验可视化），  
    对顶点进行**线性蒙皮**（Linear Blend Skinning, LBS）：
    $$
    \tilde{v} =
    \left( \sum_{j=1}^J w_j \, T_j \, B_j^{-1} \right) v_{\mathrm{rest}}(h),
    \quad
    v = \Pi(\tilde{v})
    $$
    其中 $v_{\mathrm{rest}}(h)$ 为齐次坐标，$\Pi(\cdot)$ 表示除以齐次分量 $w$ 的去齐次化。实现见 `skin_mesh/skin_character`。

## 3. Pose File 姿态与相机数据导入

首先在 Maya 中运行以下脚本，将当前帧的角色姿态和相机参数导出为 JSON：(可用于骨骼绑定或用adv插件绑定角色(需要先清除adv))
- `MayaPoseFileDataOutput.py` → `maya_posefile_output/*.json`：记录每个关节的 **局部旋转**（Joint Rotation, JR）与 **平移**。  
- `MayaPoseFileCamOutput.py` → `*_cam.json`：记录相机的 **内参矩阵**（焦距、主点等）与 **外参矩阵**（位姿）。  

1.载入姿态数据
- 读入 `pose_JR_orig`（关节局部旋转角度），格式与 `sorted_entries / local_bind` 对齐，用于重建世界矩阵。

2.生成世界矩阵（`pose_globals_orig`，调用 `process_pose.build_pose_globals`）
- 递归将 **局部绑定矩阵** 与 **局部旋转** 叠加，自根节点 $\mathrm{Root}_M$ 向下构造每个关节的世界矩阵 $G_j$：
    
    $$
    G_j =
    \begin{cases}
    L^{\mathrm{bind}}_j \cdot R_j(\mathrm{JR}), & \text{若 } \mathrm{parent}(j) = \varnothing \\
    G_{\mathrm{parent}(j)} \cdot L^{\mathrm{bind}}_j \cdot R_j(\mathrm{JR}), & \text{否则}
    \end{cases}
    $$
    
    其中 $R_j(\mathrm{JR})$ 是基于旋转顺序的欧拉旋转矩阵。

    $$
    K =
    \begin{bmatrix}
    f_x & 0 & c_x \\
    0 & f_y & c_y \\
    0 & 0 & 1
    \end{bmatrix}
    $$
    
    其中 $f_x, f_y$ 为水平/垂直焦距，$c_x, c_y$ 为主点坐标。  

- **外参矩阵** $M_{\mathrm{cam} \to \mathrm{world}}$（4×4）：描述相机在世界坐标系下的位置与朝向。  


**作用**：将 Maya 导出的关节局部旋转还原为可在世界坐标系下绘制/评分的骨架姿态，并配套相机参数，供后续投影、LoA 拟合和可视化使用。

## 4. LoA（动作线）提取与理想化

将 3D 姿态投影为 2D 关节点，自动抽取 LoA 点集，拟合 **三次 Bézier 动作线**，并通过“**视角补救 + 控制点推挤**”得到更具表现力的理想 LoA，同时返回微调后的 `pose_JR` 以供后续优化。

1.投影到 2D: `project_pose_to_2d`
- 将每个关节世界坐标 $X_w$ 变换到相机坐标系：  $X_c = R^\top (X_w - t)$, 考虑 Maya 相机朝向$(+Z → -Z)$后做透视投影，并乘以内参矩阵 $K$ 得到像素坐标 $(u,v)$。过滤相机后方点$(z \le 0)$。  

2.LoA 候选抽取: `extract_loa_candidates`
- 组装主干序列（Head → Neck → Chest → Spine → Root），用下段主干估计趋势向量 \(T\)。  
- 比较左右腿（Hip–Knee、Knee–Ankle）与 \(T\) 的夹角：
  - 若小于 `align_max_deg`（默认 35°）则视为与主干同向。  
  - 以 `knee_straight_min_deg`（默认 135°）判定“直腿”，决定是否纳入 Ankle。  
- 输出 `selected_loa`（含关节名与像素坐标）。  

3.三次 Bézier 拟合: `fit_cubic_bezier`  
- 首末点固定为 $P_0, P_3$，按弦长参数化得到 $t_i$。对$P_1, P_2$建立最小二乘方程：
$$
B(t) = (1-t)^3 P_0 + 3(1-t)^2 t P_1 + 3(1-t) t^2 P_2 + t^3 P_3, \quad\begin{bmatrix}b_1(t_i) & b_2(t_i)\end{bmatrix}
$$

- 求解 $P_1, P_2$ 以及 $t$ 列表。  

4.理想 LoA 搜索: `get_ideal_bezier(...)`，(三步)
1. **原曲线直检**：计算弯度并判别 C/S 形，若阈值达标直接返回。  
2. **视角补救**：在世界 **Y轴**对 `Root_M` 做偏航角搜索，每次重建 `pose_globals → 2D → LoA → Bézier`，选弯度更优者。  
3. **推挤控制点**：沿 $P_0 \to P_3$ 的法线方向推动 $P_1, P_2$，直到弯度达标或步长上限。若 LoA 触及腿部，则进行 **Root 等高水平平移**（保持 $v$ 不变，对齐 $u$ 到 Bézier 等高点），并将位移从像素反推到世界坐标写回 `root_pos`。

## 5. 评分模块与多阶段遗传算法优化

### 5.2 关节旋转范围定义（按阶段划分）

`OPT_JOINTS`：可优化的关节列表，每个关节对应三个旋转轴（X / Y / Z）。  

`make_ranges`：按关节集合指定旋转范围 `(min_deg, max_deg)`。  

| 阶段 | 可优化关节范围 | 主要用途 |
|------|----------------|----------|
| **Stage A** | 全身关节可动（膝盖仅绕 Z 轴） | 用于 **LoA 优化** |
| **Stage B** | 仅头部关节可动 | 优化 **朝向摄像机** |
| **Stage C** | 手臂位置回收（保持 LoA 形态不变） | 微调手臂位置 |
| **Stage D** | 腿部与下肢调整 | 优化 **silhouette + 地面接触** |
| **Stage E** | 左右手与手臂细化 | 优化 **silhouette + 接触关系** |

```python
OPT_JOINTS = ['Hip_L','Hip_R','RootPart1_M','Spine1_M','Knee_L','Knee_R','Spine1Part1_M','Ankle_L','Ankle_R',
              'Chest_M','Neck_M','Shoulder_L','Shoulder_R','NeckPart1_M','Head_M','Elbow_L','Elbow_R','Wrist_L','Wrist_R']

Larms_only  = {'Shoulder_L','Wrist_L'}
Rarms_only  = {'Shoulder_R','Wrist_R'}
legs_only   = {'Hip_L','Hip_R','Ankle_L','Ankle_R'}
Z_ONLY_leg  = {'Knee_L','Knee_R'}
Z_ONLY_arm  = {'Elbow_L','Elbow_R'}

def make_ranges(overrides, default=(0,0)):
    """overrides: list of (joint_set, [(x1,x2),(y1,y2),(z1,z2)])"""
    pr = []
    for jn in OPT_JOINTS:
        for js, rng in overrides:
            if jn in js:
                pr += rng
                break
        else:
            pr += [default, default, default]
    return pr

# 各 stage 一行定义
stageA_param_ranges = lambda: make_ranges([
    (Larms_only | Rarms_only | Z_ONLY_arm, [(0,0),(0,0),(0,0)]),
    (Z_ONLY_leg,  [(0,0),(0,0),(-10,10)]),
    (set(OPT_JOINTS), [(-30,30),(-30,30),(-30,30)])
])

stageB_param_ranges = lambda: make_ranges([
    ({'Head_M'}, [(-45,45),(-45,45),(-45,45)])
])

stageC_param_ranges = lambda: make_ranges([
    (Larms_only | Rarms_only, [(-45,45),(-45,45),(-45,45)]),
    (Z_ONLY_arm, [(0,0),(0,0),(-45,45)])
])

stageD_param_ranges = lambda: make_ranges([
    (Z_ONLY_leg, [(0,0),(0,0),(-15,90)]),
    (legs_only,  [(-90,90),(-90,90),(-90,90)])
])

stageE_param_ranges = stageC_param_ranges  # E 和 C 相同

joint_idx_map = {jn: i*3 for i, jn in enumerate(OPT_JOINTS)}
```

### 5.3 评分函数 `evaluate_pose`

输入为旋转增量向量（长度 = `len(OPT_JOINTS) × 3`），按阶段组合基准姿态后输出适应度 `fitness`：

1.应用旋转增量
- 在 `LoAfix_poseJR` 上叠加各关节 XYZ 欧拉角增量，构建新的 `pose_globals`。

2.`score.py`中调用的各项评分/惩罚的计算方式与含义

- `line_of_action_score(selected_loa, best_ctrl)`: 让 LoA 关节点序列更贴合拟合得到的三次 Bézier 曲线。做法：采样 Bézier 曲线，计算 LoA 点到曲线的最近距离均方并映射到 (0,1]。
  - 采样曲线点 $B(t)$，对每个 LoA 点 $p_i$ 取最近采样点的距离平方 $d_i^2$；
$$S_{\text{fit}}=\frac{1}{1+\lambda_{\text{fit}}\,E_{\text{fit}}}\in(0,1]$$
  - 返回 $(S_{\text{fit}}, E_{\text{fit}}, \sqrt{d_i^2})$，`agg_roothip=True` 时会把左右髋与 Root 做聚合处理，减噪更稳。

- `camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")`: 让指定关节（默认头部）面向摄像机。
  - 取头部世界位姿 $T_{\text{head}}$ 与相机世界位姿 $T_{\text{cam}}$，计算单位向量 $\hat{v}{\to \mathrm{cam}}$（从头到相机）与头部朝向向量 $\hat{v}{\text{head}}$ 的夹角余弦：
$$S_{\text{cam}}=\mathrm{clamp}_{[0,1]}(\cos\theta).$$
得分是 $\cos\theta$ 截断到 $[0,1]$，完全对向（$\theta=0^\circ$）得 1，背向（$\theta\ge 90^\circ$）得 0。

-  刚性对齐惩罚 `rigid_penalty_global_v2`: 约束若干关键关节相对参考姿势（原始或上阶段结果）的位置/朝向不要偏太多。
  - 单关节代价：
$$w_t\,\|\mathbf{t}_{\text{now}}-\mathbf{t}_{\text{ref}}\|_2 + w_r\,\angle(R_{\text{now}},R_{\text{ref}}),$$
其中旋转误差是几何角 $\angle(\cdot)$；多关节加和后取**负号**作为惩罚值返回（越负=越差）。可选对脚趾给轻锚点稳定脚掌。

- 运行时半径与 2D 胶囊 `capsule_r_px_runtime` / `build_capsules_px`: 把 T‑pose 下按权重统计得到的身体段世界半径换算成像素半径，在图像平面搭建“线段 + 半径”的 2D 胶囊，用来评估 silhouette 间距。
  - 世界半径 $;r_{\text{world}}\to$ 像素半径：
$$r_{px} = z_{\mathrm{cam}} \, f_x \, r_{\mathrm{world}}$$
锥形半径取两端投影后的较大值做保险。随后以段端点的 $(u,v)$ 和 $r_{\text{px}}$ 组装 2D 胶囊。

- Silhouette 项 `silhouette_score`: 让关键肢体与躯干之间在 2D 上保有清晰缝隙，并尽量维持 LoA 走向(`keep_ctrl`)。
  - 最小分离：对两条胶囊（或点–胶囊）求“边界最小间距”:$\mathrm{minsep} = d_{\mathrm{seg/pt}} - (r_A + r_B)$
  - 将最小间距映射为 $[-\text{overlap},1]$ 的奖励：
$$\begin{cases} \mathrm{minsep}/\text{target}, & \mathrm{minsep}\ge 0\\[2pt] \max(-\text{overlap},\,\mathrm{minsep}/(-\text{target})), & \mathrm{minsep}<0 \end{cases}$$
  - 组合项（按权重加权）：
    - 躯干–腿（只计非 LoA 腿那一侧，优先把不承载 LoA 的腿移开）
    - 躯干–臂、腿–腿、臂–臂、臂–腿
    - LoA 保持项把(`keep_ctrl` 与当前 LoA 再做一次 `line_of_action_score`)
$$w_{\text{leg\_torso}}\,\mathrm{reward}+\cdots+w_{\text{keep}}\,S_{\text{fit}}^{(\text{keep})}.$$
权重与目标缝宽（像素）可配置；远景小些、近景大些。

- 3D 自碰撞惩罚 `collision_penalty_capsule3d`: 在 3D 世界里用线段+半径的胶囊近似，抑制关键部位互相穿插。
  - 对若干成对段 $(A,B)$ 计算 3D 线段最短距 $d_{\text{3D}}$，与胶囊半径和比较：$\mathrm{sep}_{3\mathrm{D}} = d_{3\mathrm{D}} - (r_A + r_B)$
  - 若 $\mathrm{sep}_{3\text{D}}<0$ 则产生正惩罚，并按 `tgt_mm` 归一化；对最糟的 top‑k 求和，抑制噪声。数值越大越糟。

- 地面惩罚 `ground_penalty`: 对标记为“应在地面上”的脚，避免穿地并保持脚底方向合理。
  - 穿地：若脚趾/脚底 $y<y_{\text{ground}}$，按深度线性扣分；
  - 朝向：$\angle(\overrightarrow{\text{Ankle}\to\text{Toe}}, ,+Y)$ 超过阈值（默认 $25^\circ$）则扣分。

- 手贴靠惩罚/保持 `detect_hand_contacts_AB` & `hand_contact_penalty`:自动检测上一阶段 A+B 的手部贴靠锚点（如手腕贴躯干/大腿/小腿），在 E 阶段鼓励“贴在表面或略离开”，避免穿插或离太远。
  - 检测：在 AB 姿势上求手腕到目标段的分离 $\mathrm{sep}_{3\text{D}}=d-r(\cdot)$，若落在 $[-b,+b]$ 视为接触锚点；
  - E 阶段惩罚：对锚点，期望 $0\le \mathrm{sep}_{3\text{D}}\le b$，区间外按偏离量/带宽线性惩罚（穿进去重罚、离太远轻罚 `soft`）。

3.Stage 评分参数

- Stage A（LoA）$f = \mathrm{LoA\_score}(\text{selected\_loa}, \text{best\_ctrl})$
- Stage B（相机朝向 + 刚性约束） $f = 1.0 \times S_{\mathrm{cam\_head}} + 0.2 \times P_{\mathrm{rigid}}$
- Stage C（手部位置约束）$f = P_{\mathrm{rigid}}(\mathrm{Wrist\_L}, \mathrm{Wrist\_R})$
- Stage D（silhouette + 碰撞 + 地面 + 刚性约束）$f = 1.0 \times S_{\mathrm{sil}} - 0.25 \times P_{\mathrm{coll3d}} - 0.40 \times P_{\mathrm{ground}} + 0.10 \times P_{\mathrm{rigid}}$
- Stage E（silhouette + 碰撞 + 手贴靠）$f = 1.0 \times S_{\mathrm{sil}} - 0.20 \times P_{\mathrm{coll3d}} - 0.60 \times P_{\mathrm{hand\_contact}}$

```python
import score
import copy
import numpy as np

def ranges_to_mask(param_ranges):
    # 0/1 掩码：范围为 (0,0) 的位置置 0
    m = []
    for lo, hi in param_ranges:
        m.append(0.0 if (lo == 0 and hi == 0) else 1.0)
    return np.array(m, dtype=float)

def evaluate_pose(vector, stage='A', joint_order=OPT_JOINTS, base_vec=None, base_mask=None, param_ranges=None):  

    # —— 组合向量：base_vec + mask * vector ——
    vec = np.array(vector, dtype=float)
    if base_vec is None:
        composed = vec
    else:
        if base_mask is None:
            base_mask = np.ones_like(vec)
        composed = np.array(base_vec, float) + np.array(base_mask, float) * vec
        
    # 1) 深拷贝姿势，并可选Root平移
    pose_JR = copy.deepcopy(LoAfix_poseJR)

    # 2) 叠加各关节旋转增量
    for jn in joint_order:
        idx = joint_idx_map[jn]
        dx, dy, dz = composed[idx:idx+3]
        ox, oy, oz = pose_JR["joint_rot"][jn]
        pose_JR["joint_rot"][jn] = [ox+dx, oy+dy, oz+dz]

    # 3) 构建全局 & loa准备
    pose_globals = PP.build_pose_globals(sorted_entries, local_bind, parent_map, pose_JR, 'Root_M')
    points_2d    = LOAO.project_pose_to_2d(pose_globals, K, M_cam_to_world) #计算loa评分
    _, selected_loa = LOAO.extract_loa_candidates(points_2d, align_max_deg=35, knee_straight_min_deg=135)

    # 4) Stage 专属评分
    if stage == 'A':
        loa_sc, _, _ = score.line_of_action_score(selected_loa, best_ctrl)  # LoA
        fitness = (1.0 * loa_sc)
        return (float(fitness),)

    elif stage == 'B':
        cam_head = score.camera_facing_score(pose_globals, M_cam_to_world, joint_name="Head_M")
        const_head = score.rigid_penalty_global_v2(pose_globals, pose_globals_orig,
                                        constraints={'HeadEnd_M':  dict(w_t=0.005, w_r=0),},use_toe_anchor=True)   # 位置旋转约束
        fitness = (1.0 * cam_head + 0.2 * const_head)
        return (float(fitness),)

    elif stage == 'C':
        # 手位置
        const = score.rigid_penalty_global_v2(pose_globals, pose_globals_LOAO, constraints={
                                            'Wrist_L': dict(w_t=1, w_r=0.2),'Wrist_R': dict(w_t=1, w_r=0.2),},
                                             use_toe_anchor=False)   # 位置旋转约束
        fitness = (1.0 * const)
        return (float(fitness),)

    elif stage == 'D':
        r_px_by_seg = score.capsule_r_px_runtime(pose_globals, K, M_cam_to_world, radii_world) # 运行时半径
        sc_sil, sil_terms, sil_argmins, move_leg = score.silhouette_score(points_2d, r_px_by_seg,
                                                    keep_ctrl=best_ctrl, selected_loa=selected_loa,
                                                    target_gap_px=10, w_leg_torso=0.7, w_elbow_torso=0.0, w_keep=0.0,
                                                     w_arm_torso=0.0, w_leg_leg=0.15, w_arm_arm=0.0, w_arm_leg=0.0)
        # 3D：自碰撞 & 地面 & 脚旋转
        coll3d = score.collision_penalty_capsule3d(pose_globals, radii_world, topk=4, tgt_mm=0.1)
        gpen   = score.ground_penalty(pose_globals, grounded_map, ground_y, up_vec_thresh_deg=25.0)
        const = score.rigid_penalty_global_v2(pose_globals, pose_globals_LOAO, constraints={
                                            'Ankle_L': dict(w_t=0, w_r=0.2),'Ankle_R': dict(w_t=0, w_r=0.2),},
                                            use_toe_anchor=False)   # 位置旋转约束
        fitness = (1.00 * sc_sil - 0.25 * coll3d - 0.40 * gpen + 0.10 * const)
        return (float(fitness),)

    elif stage == 'E':
        r_px_by_seg = score.capsule_r_px_runtime(pose_globals, K, M_cam_to_world, radii_world) # 运行时半径
        sc_sil, sil_terms, sil_argmins, move_leg = score.silhouette_score(points_2d, r_px_by_seg,
                                                    keep_ctrl=best_ctrl, selected_loa=selected_loa,
                                                    target_gap_px=5, w_leg_torso=0.0, w_elbow_torso=0.0, w_keep=0.0,
                                                     w_arm_torso=0.2, w_leg_leg=0.0, w_arm_arm=0.2, w_arm_leg=0.15)
        # 3D：自碰撞 & 手贴靠
        coll3d = score.collision_penalty_capsule3d(pose_globals, radii_world, topk=4, tgt_mm=0.1)
        hpen = score.hand_contact_penalty(pose_globals, radii_world, hand_anchors, band_mm=0.3, soft=1.0)
        fitness = (1.00 * sc_sil - 0.20 * coll3d - 0.60 * hpen )
        return (float(fitness),)
```

### 5.4 多阶段 GA 优化流程
- 优化器: `Model.optimize_pose_GA`（基于 DEAP 框架）  
  - 种群规模：`mu` / `lambda_`, 高斯变异：`mutGaussian(sigma)`, 交叉：`cxBlend`
  - 支持阶段性的**参数冻结掩码** `mask`与种子向量 `seed_vec`


- 执行顺序
1. **Stage A** → 得到最佳 LoA 姿态向量 `bestA`  
2. **Stage B** → 在 `bestA` 基础上只动头部关节，得到 `bestB`  
3. **Stage C** → 在 `A+B` 基础上动手臂关节，得到 `bestC`  
4. **Stage D** → 在 `A+B+C` 基础上动腿部关节，得到 `bestD`  
5. **Stage E** → 在 `A+B+C+D` 基础上细化手臂，得到 `bestE`
6. 最终组合:$$base\_ABCDE = A + B + C + D + E$$

避免一次性全局搜索导致的收敛缓慢与冲突, 保证优化顺序符合动画师工作习惯：LoA → 相机 → 手臂 → 腿部 → 细化

```python
import Model
importlib.reload(Model)

# —— Stage A loa——
prA = stageA_param_ranges()
fnA = lambda v: evaluate_pose(v, stage='A', param_ranges=prA)
bestA = Model.optimize_pose_GA(fnA, prA, ngen=20, mu=30, lambda_=60, sigma=5.0, seed_vec=None)

# —— Stage B 头——
prB = stageB_param_ranges()
maskB = ranges_to_mask(prB)
fnB = lambda v: evaluate_pose(v, stage='B', base_vec=bestA, base_mask=maskB, param_ranges=prB)
bestB = Model.optimize_pose_GA(fnB, prB, ngen=20, mu=30, lambda_=60, mutpb=0.4, sigma=5.0, seed_vec=np.zeros(len(prB)), seed_spread=1.0)

# —— Stage C 手回去——
base_AB = np.array(bestA) + maskB * np.array(bestB)
prC = stageC_param_ranges()
maskC = ranges_to_mask(prC)
fnC = lambda v: evaluate_pose(v, stage='C', base_vec=base_AB, base_mask=maskC, param_ranges=prC)
bestC = Model.optimize_pose_GA(fnC, prC, ngen=30, mu=45, lambda_=80, sigma=5.0, mutpb=0.4,seed_vec=np.zeros(len(prC)), seed_spread=1.5)

# —— Stage D 腿——
base_ABC = np.array(base_AB) + maskC * np.array(bestC)
prD = stageD_param_ranges()
maskD = ranges_to_mask(prD)
fnD = lambda v: evaluate_pose(v, stage='D', base_vec=base_ABC, base_mask=maskD, param_ranges=prD)
bestD = Model.optimize_pose_GA(fnD, prD, ngen=20, mu=30, lambda_=60, mutpb=0.4, sigma=4.0, seed_vec=np.zeros(len(prD)), seed_spread=0.8)

# —— Stage E 左右手——
base_ABCD = np.array(base_ABC) + maskD * np.array(bestD)
prE = stageE_param_ranges()
maskE = ranges_to_mask(prE)
fnE = lambda v: evaluate_pose(v, stage='E', base_vec=base_ABCD, base_mask=maskE, param_ranges=prE)
bestE = Model.optimize_pose_GA(fnE, prE, ngen=20, mu=30, lambda_=60, sigma=3.0, seed_vec=np.zeros(len(prE)), seed_spread=0.8)

base_ABCDE = np.array(base_ABCD) + maskE * np.array(bestE)
```
