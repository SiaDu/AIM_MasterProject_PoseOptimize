```python
import glob, json, os

# 1. 定位到你的 JSON 文件夹
pattern = os.path.join(pose_file_path, "maya_default_pose_output/weight/verts_rest_*.json")

# 2. 批量读取并存到 dict
verts_rest_dict = {}
for path in glob.glob(pattern):
    with open(path, 'r') as f:
        data = json.load(f)
    mesh_name = data["mesh"]
    verts_rest_dict[mesh_name] = {
        "vertex_count": data["vertex_count"],
        "vertices": data["vertices"]
    }
```

```python
import glob, json, os

# 1. 定位到你的 JSON 文件夹
pattern = os.path.join(pose_file_path, "maya_default_pose_output/weight/influences_*.json")

# 2. 批量读取并存到 dict
influences_dict = {}
for path in glob.glob(pattern):
    with open(path, 'r') as f:
        data = json.load(f)
    mesh_name = data["mesh"]
    influences_dict[mesh_name] = {
        "joints": data["influences"],
        "joint_count": data["joint_count"],
        "skinCluster": data["skinCluster"]
    }
```

```python
import glob, json, os
import numpy as np

# 找到所有 weights_*.json
weights_files = glob.glob(os.path.join(pose_file_path, "maya_default_pose_output/weight/weights_*.json"))

all_weights = {}
for path in weights_files:
    data = json.load(open(path))
    mesh = data["mesh"]
    W = np.array(data["weights"])   # shape (V, J)
    all_weights[mesh] = W
```

```python
bind_inv_row = json.load(open(pose_file_path + "maya_default_pose_output/weight/bind_global_inv.json"))  # row-major
B_inv = {j: (np.array(m).reshape(4, 4)).T
         for j, m in bind_inv_row.items()}

T_dict = pose_globals_orig
```

```python
def skin_mesh(mesh):
    V_rest = np.asarray(verts_rest_dict[mesh]["vertices"])           # (V,3)
    W      = all_weights[mesh]                                       # (V,J)
    joints = influences_dict[mesh]["joints"]                         # 列顺序 ↔ 权重列

    # (J,4,4)  每根骨骼的 skin 矩阵
    skin = np.stack([T_dict[j] @ B_inv[j] for j in joints])

    V_h  = np.hstack([V_rest, np.ones((len(V_rest),1))])             # (V,4)
    VS   = (W @ skin.reshape(len(joints),16)).reshape(-1,4,4) @ V_h[...,None]
    V_def = VS[:, :3, 0] / VS[:, 3, 0][:,None]                       # 除 w
    return V_def            # 单位 cm；如需米：/100
```

```python
def skin_all_meshes(cm_to_m: bool = False):
    """返回 {mesh_name: V_def (V,3)}，可选把 cm→m"""
    out = {}
    for mesh in verts_rest_dict:
        V_def = skin_mesh(mesh)          # ← 你已写好的函数
        if cm_to_m:
            V_def = V_def / 100.0
        out[mesh] = V_def
    return out

V_all_dict = skin_all_meshes(cm_to_m=False)     # 统一转成米
# -------- ① 拼所有 mesh 顶点（示例，假设已转米） --------
V_concat = np.concatenate(list(V_all_dict.values()), axis=0)   # (N,3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



# -------- ② 画 3D 散点（等比例） --------
fig = plt.figure(figsize=(6, 6))
ax  = fig.add_subplot(111, projection="3d", box_aspect=(1, 1, 1))
ax.scatter(V_concat[:, 0], V_concat[:, 1], V_concat[:, 2], s=0.3)

# --- 等比例范围 ---
# 用 np.ptp 兼容 NumPy 2.0+
range_xyz = np.ptp(V_concat, axis=0)        # 每个轴的最大–最小
max_range = range_xyz.max() / 2
mid       = V_concat.mean(axis=0)

for setter, m in zip([ax.set_xlim, ax.set_ylim, ax.set_zlim], mid):
    setter(m - max_range, m + max_range)

ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
ax.set_zlabel("Z (m)")
plt.show()
```
