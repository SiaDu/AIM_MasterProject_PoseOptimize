从 Maya 导出的 joints_hierarchy.json 中，递归计算每个 joint 的世界变换矩阵（Tpose），并提取其世界位置（joint 的 pivot 位置）

````python
file_path = "/home/s5722875/Notes/Semester3/SkeletonFinished/maya_default_pose_output/"
````

````python
import numpy as np
import json

# 读取 JSON 文件
with open(file_path + "joints_hierarchy_local.json", "r") as f:
    j_hierarchy = json.load(f)
````

Maya的`cmds.xform(..., m=True) `返回的是按行序展开的 16 个数,
```
[ Xx  Xy  Xz   0 ]   ← 每行是方向/位移向量
[ Yx  Yy  Yz   0 ]
[ Zx  Zy  Zz   0 ]
[ Tx  Ty  Tz   1 ]
```
结构为`[x, y, z, 1] × M = new_vec`

`local_bind[joint_name]`：每个关节的局部绑定矩阵（$4\times4$，表示该关节相对于其 parent 的矩阵）
`parent_map[joint_name]`：每个关节的 parent 是谁（短名字）

````python
parent_map = {}
for entry in j_hierarchy:
    short_name = entry["name"].split("|")[-1]
    parent = entry["parent"]
    parent_map[short_name] = parent.split("|")[-1] if parent else None

depth_map = {}
def get_depth(name):
    if parent_map[name] is None:
        return 0
    if name in depth_map:
        return depth_map[name]
    d = 1 + get_depth(parent_map[name])
    depth_map[name] = d
    return d

sorted_entries = sorted(
    j_hierarchy,
    key=lambda e: get_depth(e["name"].split("|")[-1])
)

local_bind = {}
joint_orient_map = {}
for entry in sorted_entries:
    short_name = entry["name"].split("|")[-1]
    # Maya reshape row-major 列表到 4×4 标准齐次矩阵
    local_bind[short_name] = np.array(entry["local_matrix"]).reshape((4,4), order='F')
    joint_orient_map[short_name] = entry["rotateOrder"].lower() # .lower()确保都是小写
````

递归计算世界变换矩阵（global transform）

$$
T^{j}_{\mathrm{world}}
\;=\;
T^{\mathrm{parent}}_{\mathrm{world}}
\;\cdot\;
T^{j}_{\mathrm{local}}
$$


递归：`compute_bind_global(p)` 会先一直向上走到根节点，拿到根的矩阵，然后一路 “父 @ 子” 地往下累乘回来，最终给出 j 的全局矩阵。

矩阵乘法顺序：Parent_global @ Local_j，保持了从根到叶子的变换链。

````python
def compute_bind_global(j):
    L = local_bind[j]  # 1. 拿到关节 j 的 “局部绑定矩阵” L（4×4）
    p = parent_map[j]  # 2. 查它的父关节名 p（如果是根节点，则 p 为 None）
    return L if p is None else compute_bind_global(p) @ L
    # 3a. 如果是根节点，就直接把它的局部矩阵当作全局矩阵返回
    # 3b. 否则，先递归算出父关节的全局矩阵，
    #     再乘以自己（L），得到自己的全局矩阵
````

````python
global_bind = {j: compute_bind_global(j) for j in local_bind}
````

````python
import transforms3d as t3d
from transforms3d import affines    # 用来一行生成齐次矩阵

# —— 辅助：把度转成弧度
def deg2rad(v):
    return [x * np.pi/180.0 for x in v]
````

````python
# —— 读取该 Pose 的 Euler + root旋转平移 ——

#pose = json.load(open(pose_file_path + "G1_happy_bad_front_1.json", "r"))
#pose = json.load(open(pose_file_path + "G2_stand_bad_front_1.json", "r"))
#pose = json.load(open(pose_file_path + "G3_CrouchDownAndPickUp_bad_front_1.json", "r"))
#pose = json.load(open(pose_file_path + "G4_SadStand_bad_front_1.json", "r"))
pose = json.load(open(pose_file_path + "G5_AngryPunch_bad_front_1.json", "r"))

# —— 用 pose 数据直接生成根节点世界变换（包含平移+旋转）
root_pos     = np.array(pose["root_pos"])        # cm
root_rot_deg = pose["root_rot"]                  # [rx,ry,rz] 单位度
root_rot_rad = deg2rad(root_rot_deg)             # 转成弧度
R_root_p     = t3d.euler.euler2mat(*root_rot_rad, axes='sxyz')
world_mats   = {
    "Root_M": affines.compose(root_pos, R_root_p, [1,1,1])
}
````

````python
print(root_pos)
print(root_rot_deg)
print(root_rot_rad)
print(R_root_p)
print(world_mats)
````

````python
def build_pose_globals(sorted_entries, local_bind, parent_map, pose):
    world = dict(world_mats)   # 先把 Root_M 带进去

    for entry in sorted_entries:
        name   = entry["name"].split("|")[-1]
        if name == "Root_M":
            continue         # Root 已经处理过了

        # 父节点名字
        parent = parent_map[name] or "Root_M"

        # 本地 bind 矩阵
        B = local_bind[name]              # shape (4,4)
        t_bind = B[:3, 3]                 # bind 平移
        R_bind = B[:3, :3]                # bind 旋转

        # pose 的局部 Euler（度），如果没指定就全 0
        eul_deg = pose["joint_rot"].get(name, [0.0, 0.0, 0.0])
        eul_rad = deg2rad(eul_deg)        # 转成弧度
        R_p     = t3d.euler.euler2mat(*eul_rad, axes='rxyz')

        # 先 bind 再 pose
        R_new   = R_bind @ R_p

        # 组装成 4×4 的 local_P
        local_P = np.eye(4)
        local_P[:3, :3] = R_new
        local_P[:3, 3]  = t_bind

        # 递归叠加到父节点的 world 上
        world[name] = world[parent] @ local_P

    return world
````

````python
pose_globals = build_pose_globals(
    sorted_entries,
    local_bind,
    parent_map,
    pose
)
````
